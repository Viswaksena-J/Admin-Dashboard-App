```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(-10,10,0.1)
```

```python
from scipy import signal
from scipy.signal import square
```

```python
y = signal.square(x)
```

```python
plt.plot(x,y)
```

```python
x = np.repeat([0.,1.,0.],100)
plt.plot(x)
```

```python
from scipy.signal import sawtooth,unit_impulse
# x = np.linspace(0,1,1000,endpoint=False)
x = np.arange(0,6*np.pi,0.1)
y = signal.sawtooth(x)
plt.plot(x,y)
```

```python
# x = np.linspace(0,1,1000,endpoint=False)
y = signal.unit_impulse(len(x),'mid')
plt.plot(x,y)
```

```python
y = np.heaviside(x-0.5,1)
plt.plot(x,y)
```
